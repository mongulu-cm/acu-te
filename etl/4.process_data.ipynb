## Necessary imports

```python
import numpy as np
import pandas as pd
```

## Load and viz data

```python
data = pd.read_csv("./ref-rna-real-mars-2022-enriched-not-qualified.csv", index_col=0)
data = data[data.columns[1:]] # ignore first column it is index not correctly saved
```

```python
data.info()
```

```python
data.describe(include=object)
```

```python
data.sample(5)
```

## Title in uppercase and objet in lower

```python
data["titre"] = data["titre"].apply(lambda x: str(x).upper())
data["objet"] = data["objet"].apply(lambda x: str(x).lower())
```

```python
data.sample(5)
```

## Remove unecessary spaces

```python
text_columns = [
    "titre", "objet", "social_object1_libelle", "social_object2_libelle"
]
```

```python
def remove_space_at_the_end(x: str):
    if x is not None:
        return x.strip()
```

```python
for column in text_columns:
    data[column] = data[column].apply(remove_space_at_the_end)
```

```python
data.sample(5)
```

## Remove extra "

```python
data[data["id"] == "W751228313"]
```

```python
data[data["id"] == "W751228313"]["objet"].values
```

```python
def replace_double_quote(x: str):
    if x is not None:
        return x.replace("\"\"", "'")
```

```python
data["objet"] = data["objet"].apply(replace_double_quote)
```

```python
print(data[data["id"] == "W751228313"]["objet"].values[0])
```

## Save without index

```python
filename = 'ref-rna-real-mars-2022-enriched-not-qualified-process'
compression_options = dict(method='zip', archive_name=f'{filename}.csv')
data.to_csv(f'./{filename}.zip', compression=compression_options, index=False)
```
